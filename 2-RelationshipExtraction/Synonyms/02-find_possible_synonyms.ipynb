```python
import dill
from gensim.models import Word2Vec
import pandas as pd
from ast import literal_eval
import csv
```

## Get All the Words in Corpus

```python
#read in data and set dictionary values as dictionaries
file = open("clean_sentences.csv", "r")
csv_reader = csv.reader(file)

sentences = []
for row in csv_reader:
    sentences.append(row)
```

```python
all_words = []
for doc in sentences:
    for word in doc:
        all_words.append(word)
words = set(all_words)
```

```python
print(len(words))
```

## Load the Embeddings Model that we Trained

```python
modelObj = dill.load(open("pickled_models/full_corpus.pkl", "rb"))
```

```python
# run through all the words and find the synonyms given a particular threshold (0.8)
synonyms = []
count = 0
for word in words:
    try:
        similar_words = modelObj.model.wv.most_similar(positive=[word], topn = 10, restrict_vocab=None)
        for w in similar_words:
            if w[1] > 0.8:
                synonyms.append([word, w[0]])
    except:
        pass
    count += 1
    if count % 100 == 0:
        print("Through {} words".format(count))
```

```python
print(len(synonyms))
```

```python
def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)
```

```python
has_numbers('1')
```

```python
# take out numbers
clean_synonyms = []
for pair in synonyms:
    if has_numbers(pair[0]) or has_numbers(pair[1]):
        pass
    elif len(pair[0]) == 1 or len(pair[1]) == 1:
        pass
    else:
        clean_synonyms.append(pair)
```

```python
print(len(clean_synonyms))
```

```python
sorted_syns = []
for pair in clean_synonyms:
    new_list = sorted(pair, reverse=False)
    sorted_syns.append(new_list)
```

```python
final_synonyms = set(tuple(row) for row in sorted_syns)
```

```python
print(len(final_synonyms))
```
