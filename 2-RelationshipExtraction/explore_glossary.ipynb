```python
import pandas as pd
from ast import literal_eval
```

```python
#read in data and set dictionary values as dictionaries
df = pd.read_csv('full_dataframe.csv', index_col=0)
df['doc_type'].replace('admin%20instructions', 'admin instructions', inplace=True)
df['acronyms'] = df['acronyms'].apply(literal_eval)
df['glossary'] = df['glossary'].apply(literal_eval)
```

```python
# get dataframe of documents with glossary, and dataframe of the documents that had glossary terms extracted
docs_w_glossary = df[df['raw_text'].apply(lambda x: 'glossary' in x.lower())]
docs_gloss_extracted = df.loc[df['glossary'].apply(lambda x: len(x)>1)]
```

```python
print("{} documents with glossaries".format(len(docs_w_glossary)))
print("{} were picked up with the getGlossary() function".format(len(docs_gloss_extracted)))
print("That's {:.0%} percent.".format(round(len(docs_gloss_extracted)/len(docs_w_glossary),2)))
```

```python
#Explore by document type
docs_w_glossary.groupby("doc_type").size().reset_index(name="count")
```

```python
#Explore by document type
docs_gloss_extracted.groupby("doc_type").size().reset_index(name="count")
```

```python
#calculate percentage extracted from each type
percent_list = list(docs_gloss_extracted.groupby("doc_type").size().reset_index(name="count")['count']/docs_w_glossary.groupby("doc_type").size().reset_index(name="count")['count'])
as_percentages_list = [ str(round(x*100,1))+"%" for x in percent_list]
```

```python
# get percentage extracted
percent_dict = {
    'doc_type': ['admin instructions', 'directives', 'instructions', 'manuals', 'memos'],
    'percent': as_percentages_list
}
percent_df = pd.DataFrame(percent_dict)
percent_df.head()
```

```python
total_words = 0
for index, row in docs_gloss_extracted.iterrows():
    print("\033[1m Doc Type: {} \033[0m".format(row['doc_type']))
    print("\033[1m File Name: {} \033[0m".format(row['file_name']))
    for key, value in row['glossary'].items():
        print(key + ":")
        print(value)
        total_words+=1
        print("\n")
```

```python
print("There are {} glossary words total".format(total_words))
```
