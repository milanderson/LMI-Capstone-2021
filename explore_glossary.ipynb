```python
import pandas as pd
from ast import literal_eval
```

```python
df = pd.read_csv('full_dataframe.csv', index_col=0)
df['acronyms'] = df['acronyms'].apply(literal_eval)
df['glossary'] = df['glossary'].apply(literal_eval)
```

```python
df_g = df.loc[df['glossary'].apply(lambda x: len(x)>1)]
```

```python
for index, row in df_g.iterrows():
    for key, value in row['glossary'].items():
        print(key + ":")
        print(value)
        print("\n")
```
