```python
import pandas as pd
from ast import literal_eval
```

```python
# read in synonyms
synonyms_df = pd.read_csv("final_synonyms.csv", index_col = 0)
```

```python
synonyms_df.head()
```

```python
# read in corpus
corpus_df = pd.read_csv("full_custom_tokenized_df.csv", index_col = 0)
corpus_df['doc_type'].replace('admin%20instructions', 'admin instructions', inplace=True)
```

```python
synonyms = []
bigrams = []
count = 0
for index, row in synonyms_df.iterrows():
    count += 1
    for text in corpus_df['cleaned_text']:
        if row['First Term'] + " " + row['Second Term'] in text.lower() or row['Second Term'] + " " + row['First Term'] in text.lower():
            print("Bigram found:" + row['First Term'] + " " + row['Second Term'])
            bigrams.append((row['First Term'], row['Second Term']))
            break
        else:
            continue
    synonyms.append((row['First Term'], row['Second Term']))
    if count % 10 == 0:
        print("Done with {}.".format(count))
```

```python
for bigram in bigrams:
    synonyms.remove(bigram)
```

```python
synonyms
```

```python
df = pd.DataFrame(synonyms, columns = ['First Term', 'Second Term'])
```
